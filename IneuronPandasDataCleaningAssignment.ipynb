```python
import pandas as pd
import numpy as np
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
df
```

```python
for i in df.loc[df['FlightNumber'].isnull()].index:
    df['FlightNumber'][i] = df['FlightNumber'][i-1]+10
print(df['FlightNumber'])
df
```

```python
From = []
To = []
for i in range(len(df['From_To'])):
    sep_from_to = df['From_To'][i].split(sep='_')
    for j in range(len(sep_from_to)):
        if j == 0:
            From.append(sep_from_to[j])
        else:
            To.append(sep_from_to[j])

            
            
df_from_to = pd.DataFrame({'From': From , 'To': To})
df_from_to
```

```python
for i in range(len(df_from_to)):
    for j in range(len(df_from_to.iloc[i])):
        df_from_to.iloc[i,j] = df_from_to.iloc[i,j].capitalize() 
df_from_to
```

```python
df = df.drop('From_To',axis=1)
df = pd.concat([df_from_to,df],axis=1)
df
```

```python
df[['delays_1','delays_2','delays_3']] = pd.DataFrame(df.RecentDelays.tolist(), index= df.index)
df
```

```python
df = df.drop('RecentDelays',axis=1)
df
```

```python

df['Airline'] = df['Airline'].str.extract('([a-zA-Z ]+)',expand=False)
df
```
